```python
import pandas as pd
import ast
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('hh.csv')
```

```python
plt.figure(figsize=(10,7))
plt.pie(df.employer.value_counts()[:25], shadow=True);
plt.legend(labels=df.employer.value_counts()[:25].index.tolist(), bbox_to_anchor=(1,1));

hole = plt.Circle((0, 0), 0.65, facecolor='white');
plt.gcf().gca().add_artist(hole);
```

```python
from wordcloud import WordCloud
cloud = ''
for x in df.employer.value_counts()[:25].index.tolist():
    cloud += x + ' '
wordcloud = WordCloud(width = 800, height = 800,
                min_font_size = 8, background_color='white'
                     ).generate(cloud)
```

```python
plt.figure(figsize = (20,10)) 
plt.imshow(wordcloud);
plt.savefig('vacancy_cloud.png')
```
