```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_palette('pastel')
import warnings
warnings.filterwarnings('ignore')
```

```python
df_before = pd.read_csv('student_before_score.csv')
df_after = pd.read_csv('student_after_score.csv')
df_group = pd.read_csv('student_group.csv')
```

```python
df_before.isnull().sum(), df_after.isnull().sum()
```

```python
df_before.shape
```

```python
df = pd.merge(df_before, df_after, on='student_id')
df = pd.merge(df, df_group, how='left', on='student_id')
```

```python
df.head()
```

- Тысяча учеников, нет отсутствующих значений

## Visualization ##

```python
fig, ax = plt.subplots(1, 2, figsize=(10,4), sharey=True)
sns.boxplot(df[['group', 'before_score']], x='group', y='before_score', ax=ax[0]);
sns.boxplot(df[['group', 'after_score']], x='group', y='after_score', ax=ax[1]);

ax[0].set_ylabel('Балл');
ax[0].set_xlabel('Пойдет ли на подготовительный курс');
ax[0].set_title('До прохождения курса');

ax[1].set_ylabel('Балл');
ax[1].set_xlabel('Прошел ли подготовительный курс');
ax[1].set_title('После прохождения курса');
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10,4), sharey=True)

sns.distplot(df[df['group'] == 'course']['before_score'], color='skyblue', ax=ax[0]);
sns.distplot(df[df['group'] == 'no_course']['before_score'], color='red', ax=ax[0]);
ax[0].set_xlabel('Пойдет ли на подготовительный курс');
ax[0].set_title('До прохождения курса');

sns.distplot(df[df['group'] == 'course']['after_score'], color='skyblue', ax=ax[1]);
sns.distplot(df[df['group'] == 'no_course']['after_score'], color='red', ax=ax[1]);
ax[1].set_xlabel('Прошел ли подготовительный курс');
ax[1].set_title('После прохождения курса');
```

## Анализ групп ##

Проверим, является ли распредление оценок нормальным
- Если является, проведем **Т-тест**
- Если нет, то тест **Манна-Уитни**

Для проверки на нормальность проведем **тест Шапиро-Вилка**

```python
from scipy.stats import shapiro, levene
```

- Нулевая гипотеза: у переменной нормальное распределение, p-value > 0.05
- Альтернативная гипотеза: у переменной ненормальное распределение, p-value < 0.05

```python
stat, p_value = shapiro(df['before_score'] )
print(f'Статистика: {stat}, p-value: {p_value}')

stat, p_value = shapiro(df['after_score'])
print(f'Статистика: {stat}, p-value: {p_value}')
```

- Отвергаем нулевую гипотезу, у переменной ненормальное распределение, проведем тест Манна-Уитни

## Тест Манна-Уитни ##

- Нулевая гипотеза: между группами нет значительной разницы, p-value > 0.05
- Альтернативная гипотеза: есть значительная разница, p-value < 0.05

### До прохождения подготовительных курсов ###

```python
from scipy.stats import mannwhitneyu
u_value, mannw_test_p = mannwhitneyu(df[df['group'] == 'course']['before_score'], df[df['group'] == 'no_course']['before_score'])
print(f'U-value: {u_value}, p-value: {round(mannw_test_p, 2)}')
```

- Мы не можем отвергнуть нулевую гипотезу, значит, между группами нет значительной разницы.
- Следовательно, ученики в двух группах ничем не различаются, кроме прохождения подготовительных курсов.

### После прохождения ###

```python
u_value, mannw_test_p = mannwhitneyu(df[df['group'] == 'course']['after_score'], df[df['group'] == 'no_course']['after_score'])
print(f'U-value: {u_value}, p-value: {round(mannw_test_p, 2)}')
```

- Отвергаем нулевую гипотезу, после прохождения подготовительных курсов между оценками двух групп наблюдается значительная разница
- Соответсвенно, **посещение курсов повлияло на балл студентов между группами**
